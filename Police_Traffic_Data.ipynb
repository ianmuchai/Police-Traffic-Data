```python
import pandas as pd
```

```python
df = pd.read_csv("/content/3. Police Data.csv")
df
```

```python
# Locating and removing columns with missing values
df.isnull().sum()
```

```python
df.drop(columns = 'country_name', inplace = True)
#Column has been successfully removed from the data
```

```python
# Checking for speeding to see which gender was stopped more often that the other, male or female.
df[df.violation == 'Speeding'].driver_gender.value_counts() #Using filter function and value_counts to meet requirements.
```

```python
# Does gender affect whether a person is searched during a stop?
df.groupby('driver_gender').search_conducted.sum() # Group by gender and find summation of search_conducted instances.
```

```python
df.search_conducted.value_counts()
```

```python
# Finding the mean stop_duration by mapping and data type casting
df.stop_duration.value_counts()
```

```python
df['stop_duration'] = df['stop_duration'].map({'0-15 Min' : 7.5 , '16-30 Min' : 24 , '30+ Min' : 45 })
df
```

```python
df['stop_duration'].mean()
```

```python
# Comparing age distribution for each violation
df.groupby('violation').driver_age.describe()
```
